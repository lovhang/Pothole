```python
mdl = Model('pothole')
x = mdl.binary_var_dict([(i,j) for i in n for j in t], name='x')
```

```python
mdl.maximize(mdl.sum(w[i][j]*x[i,j] for i in n for j in t))
mdl.add_constraints(mdl.sum(l[i][j]*x[i,j] for i in n) <= L for j in t)
```

```python
solution = mdl.solve(log_output=True)
```
