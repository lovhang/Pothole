```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(f'C:\\users\\xiaohang\\Documents\\roadSimple.csv')
```

```python
df = df.dropna(subset = ['id_road'])
```

```python
df
```

```python
df['id_road'] = df['id_road'].astype(np.int64)
```

```python
max_num = df['id_road'].max()
max_num
```

```python
max_num = int(max_num)
max_num
```

```python
cl = [0 for i in range(0,max_num+1)]
```

```python
cl
```

```python
for i, row in df.iterrows():
    cl[row['id_road']] = cl[row['id_road']] + 1
```

```python
cl
```

```python
max(cl)
```

```python
for idx, val in enumerate(cl):
    if val > 1:
        print(idx)
        
```
